# 리스트 컴프리헨션 (List Comprehension)

## 개요
기존 리스트를 기반으로 새로운 리스트를 만들어내는 파이썬만의 간결하고 강력한 문법입니다.

## 장점
- **간결성**: 코드가 짧고 직관적입니다.
- **성능**: 일반적인 `for` 루프보다 실행 속도가 빠릅니다 (내부 최적화).

## 기본 문법
```python
[표현식 for 항목 in 반복가능객체 if 조건문]
```

````python
print("=== 1. 기본 형태 ===")
# [표현식 for 항목 in 반복가능객체]
squares = [x**2 for x in range(10)]
print(f"0~9 제곱 리스트: {squares}")

print("\n=== 2. 조건문 포함 ===")
# [표현식 for 항목 in 반복가능객체 if 조건문]
evens = [x for x in range(10) if x % 2 == 0]
print(f"0~9 짝수 리스트: {evens}")

print("\n=== 3. 2차원 배열 초기화 (코테 필수) ===")
# 0으로 채워진 3x3 행렬 만들기
# 주의: [[0] * 3] * 3 으로 만들면 얕은 복사 문제가 발생할 수 있음
matrix = [[0 for _ in range(3)] for _ in range(3)]
print(f"3x3 Matrix: {matrix}")

print("\n=== 4. 문자열 처리 활용 ===")
words = ["apple", "banana", "cherry"]
upper_words = [w.upper() for w in words]
print(f"Uppercased: {upper_words}")
````
